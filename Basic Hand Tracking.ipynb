```python
import cv2
import mediapipe as mp
import time

# Set input camera to first (in my case only camera) plugged in
cap = cv2.VideoCapture(0)

# Bring in Hand Tracking utils
mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils

pTime = 0
cTime = 0

# While true, run camera
while True:
    success, img = cap.read()
    
    # Convert to RGB
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Check for hands
    results = hands.process(imgRGB)
    #print(results.multi_hand_landmarks)
    
    # Draw landmarks if detected
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            # Get Location for each landmark on hand
            for id, lm in enumerate(handLms.landmark):
                #print(id, lm)
                # Get Height, Width, Channel of camera output
                h, w, c = img.shape
                # Multiply landmark decimal location by w/h to get pixel location
                cx, cy = int(lm.x * w), int(lm.y * h)
                #print
                print(id, cx, cy)
                # Draw a cirlce around a specific landmark, 4 = tip of thumb
                if id == 4:
                    cv2.circle(img, (cx, cy), 15, (255, 0, 255), cv2.FILLED)
            
            mpDraw.draw_landmarks(img, handLms, mpHands.HAND_CONNECTIONS)
            
    # Calc fps and display
    cTime = time.time()
    fps = 1 / (cTime-pTime)
    pTime = cTime
    cv2.putText(img, str(int(fps)), (10, 70), cv2.FONT_HERSHEY_PLAIN, 3, (255, 0, 0), 3) 
    
    # Show Camera Image
    cv2.imshow("Image", img)
    cv2.waitKey(1)
```
